```python

barrios_top_10 = ["Usaquén", "Zona Noroccidental", "Bosa", "Kennedy", "Cedritos", "Barrios Unidos", "Engativá", "Suba", "Santa Bárbara", "Chapinero"]
inmuebles_top_10 = df[df['Barrio'].isin(barrios_top_10)]

# Calcular el promedio del área
promedio_area = inmuebles_top_10['Area'].mean()

# Calcular otros datos estadísticos
conteo = inmuebles_top_10['Area'].count()
mediana = inmuebles_top_10['Area'].median()
minimo = inmuebles_top_10['Area'].min()
maximo = inmuebles_top_10['Area'].max()

# Imprimir los resultados
print("Promedio del área:", promedio_area)
print("Conteo:", conteo)
print("Mediana:", mediana)
print("Valor mínimo:", minimo)
print("Valor máximo:", maximo)
```

```python

inmuebles_barrio.head(10).plot.bar(x='Barrio', y='Area', figsize=(10, 6))
plt.title('Área de Inmuebles en los 10 Barrios Principales')
plt.xlabel('Barrio')
plt.ylabel('Área')
plt.xticks(rotation=45)
plt.show()
```
